# Data preprocessing tools

## Importing the library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy= 'mean')
```

```python
imputer.fit(X[:, 1:3])      # look for all the missing values in specified columns
X[:, 1:3] = imputer.transform(X[:, 1:3])    # now replace the missing values in a column with mean of the values in that column
```

```python
print(X)  # nan values have been replaced!
```

## Encoding categorical data

we'll be using 'one hot encoding' method (binary vectors of each category)

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])] , remainder = 'passthrough')   # passthrough means we want to keep the columns that we don't apply the transformation on
X = np.array(ct.fit_transform(X))   # applying the transformation and updating 'X', here we use np.array b/c fit_transform doesn't return np array so we're explicitly converting the output
```

```python
print(X)     # the country column is gone and there are 3 columns, France is 1,0,0 each value is in different column
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()    # no parameters as it's only one single vector we want to apply it on
y = le.fit_transform(y)    # this doesn't need to be a np array
```

```python
print(y)   # O is No, 1 is yes
```
